```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.array([(1, 1), (2, 4), (3, 2), (7, 5)])
# get x and y vectors
x = points[:,0]
y = points[:,1]
```

```python
# calculate polynomial
z = np.polyfit(x, y, 3)
print(z)
```

```python
f3 = np.poly1d(z)
print(f3)
```

```python
# calculate new x's and y's
x_new = np.linspace(x[0], x[-1], 50)
y_new_3 = f3(x_new)
```

```python
plt.plot(x,y,'o', x_new, y_new_3)
plt.xlim([x[0]-1, x[-1] + 1 ])
plt.show()
```
